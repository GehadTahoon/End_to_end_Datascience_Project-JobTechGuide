##### constants

```python
DATA_PATH = '../data/processed/1_preprocessed_data.pkl'
ROLES_PATH = '../data/raw/roles_short_names.csv'


ROLE_COLS = ['DevType']
TECH_COLS = [
    'LanguageHaveWorkedWith',
    'DatabaseHaveWorkedWith',
    'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'MiscTechHaveWorkedWith',
    'ToolsTechHaveWorkedWith',
    'NEWCollabToolsHaveWorkedWith'
]
```

##### load packages

```python
import pandas as pd 
import numpy as np
import logging
import pickle


from sklearn.preprocessing import MultiLabelBinarizer
```

##### _______________________________________________________________________________________________________________________

##### Read data

```python
raw_data = pd.read_pickle(DATA_PATH)
role_names = pd.read_csv(ROLES_PATH, sep=';')
```

##### encoding


```python
data = raw_data.copy()
encoded_data ={}
for col in ROLE_COLS + TECH_COLS:
    binarizer = MultiLabelBinarizer()   
    encoded_df = pd.DataFrame(binarizer.fit_transform(data[col]),
                              columns=binarizer.classes_,
                              index=data[col].index
                             )   
    encoded_data[col] = encoded_df


#    This is typically used when dealing with columns that contain lists of items (like roles or technologies someone knows),
# and you want to convert these lists into a binary format (one-hot encoding) for easier analysis.
```

```python
data = pd.concat(encoded_data, axis=1)
data
```
